## **	Lists problems**

Calculate the sum, count of even numbers, odd numbers from a list

```python
numbers = [15, 16, 20, 3, 9, 20]


total_sum = sum(numbers)

even_count = 0
odd_count = 0
for number in numbers:
    if number % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print(f"List: {numbers}")
print(f"Sum of the list: {total_sum}")
print(f"Count of even numbers: {even_count}")
print(f"Count of odd numbers: {odd_count}")
```

Check if an element exists in a list in Python  if exists print its index

```python
lst = [3, 6, 3, 5, 3, 4]
num = 7
if num in lst:
    index = lst.index(num)
    print(f"{num} is found on index = {index}")
else:
    print("number is not found")

lst = [1, 6, 3, 5, 3, 4]
num = 6
if num in lst:
    index = lst.index(num)
    print(f"{num} is found on index = {index}")
else:
    print("number is not found")
```
