# Lab 4 OOP Pratice

```python
class Customer:
      def __init__(self,customer_id,customer_name,customer_phone,customer_address):
        self.customer_id=customer_id
        self.customer_name=customer_name
        self.customer_phone=customer_phone
        self.customer_address=customer_address
```

```python
class Company(Customer):
  def __init__(self,customer_id,customer_name,customer_phone,customer_address,contact,discount):
    super().__init__(customer_id,customer_name,customer_phone,customer_address)
    self.contact=contact
    self.discount=discount
```

```python
class Individual(Customer):
  def __init__(self,customer_id,customer_name,customer_phone,customer_address,lic_number):
    super().__init__(customer_id,customer_name,customer_phone,customer_address)
    self.lic_number=lic_number
```

```python
class Product:
  def __init__(self,product_id,product_name,product_retail_price,product_description):
    self.product_id=product_id
    self.product_name=product_name
    self.product_retail_price=product_retail_price
    self.product_description=product_description
```

```python
class Software(Product):
  def __init__(self,product_id,product_name,product_retail_price,product_description,license):
    super().__init__(product_id,product_name,product_retail_price,product_description)
    self.license=license
```

```python
class Manual(Product):
  def __init__(self,product_id,product_name,product_retail_price,product_description,publisher):
    super().__init__(product_id,product_name,product_retail_price,product_description)
    self.publisher=publisher
```

```python
class Taxable:
  def __init__(self,tax_percent):
    self.tax_percent=tax_percent
```

```python
class Hardware(Product,Taxable):
  def __init__(self,product_id,product_name,product_retail_price,product_description,tax_percent,warranty_period):
    Product.__init__(self, product_id, product_name, product_retail_price, product_description)
    Taxable.__init__(self, tax_percent)
    self.warranty_period=warranty_period
```
