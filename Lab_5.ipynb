### Vacation Package

```python
from typing import List
from abc import abstractmethod
```

```python
class VacationPackage :
    def __init__(self,destination,price):
        self.destination = destination
        self.price = price

    def display_ditails(self):
        print(f"Destination: {self.destination}")
        print(f"Price: {self.price}")
```

```python
class RelaxationPackage(VacationPackage):
  def __init__(self,destination,price,Spa_Service: List):
    super().__init__(destination,price)
    self.Spa_Service = Spa_Service

  def display_ditails(self):
    super().display_ditails()
    print(f"Spa Service: {self.Spa_Service}")
```

```python
class AdventurePackage(VacationPackage):
  def __init__(self,destination,price,Sea_Service: List):
    super().__init__(destination,price)
    self.Sea_Service = Sea_Service

  def display_ditails(self):
    super().display_ditails()
    print(f"Adventure Service: {self.Sea_Service}")
```

```python
Adventure=AdventurePackage('Turkey',1000,['Hiking','Swimming','Fishing'])
```

```python
Adventure.display_ditails()
```

```python
Relaxation=RelaxationPackage('Egypt',3000,['Spa','Massage'])
Relaxation.display_ditails()
```

### Hotel Tour

```python
class TravelPackage:
  @abstractmethod
  def calc_total_cost(self):
    pass
  @abstractmethod
  def display_total_cost(self):
    pass
```

```python
class HotelPackage(TravelPackage):
  def __init__(self,NumOfNights,CostPerNight):
    self.NumOfNights = NumOfNights
    self.CostPerNight = CostPerNight
  def calc_total_cost(self):
    return self.NumOfNights * self.CostPerNight
  def display_total_cost(self):
    print(f"Total Cost for hotel: {self.calc_total_cost()}")
```

```python
class TourPackage(TravelPackage):
  def __init__(self,NumOfDays,CostPerDay):
    self.NumOfDays = NumOfDays
    self.CostPerDay = CostPerDay
  def calc_total_cost(self):
    return self.NumOfDays * self.CostPerDay
  def display_total_cost(self):
    print(f"Total Cost for tour: {self.calc_total_cost()}")
```

```python
HotelPac=HotelPackage(5,100)
HotelPac.calc_total_cost()
HotelPac.display_total_cost()
```

```python
TourPac=TourPackage(10,100)
TourPac.calc_total_cost()
TourPac.display_total_cost()
```

### Patient Doctor Model

```python
class Person :
  def __init__(self,name,age):
    self.name = name
    self.age = age
```

```python
class Patient(Person):
  def __init__(self,name,age,medicial_history):
    super().__init__(name,age)
    self.medicial_history = medicial_history
  def display_details(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Medicial History: {self.medicial_history}")
```

```python
class Doctor(Person):
  def __init__ (self,name,age,spcalization):
    super().__init__(name,age)
    self.spcalization = spcalization
  def display_details(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Spcalization: {self.spcalization}")
```

```python
def print_detials(person):
  person.display_details()
```

```python
Ahmed=Patient('Ahmed',20,'Cold')
Mohamed=Doctor('Mohamed',30,'Heart')
```

```python
print_detials(Ahmed)
print_detials(Mohamed)
```
